```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv(r'C:\Users\MOHITH\OneDrive\Desktop\STUDY MATERIALS\ML\Machine Learning A-Z (Codes and Datasets)\Part 2 - Regression\Section 6 - Polynomial Regression\Position_Salaries.csv')
X=df.iloc[:,1:-1].values
y=df.iloc[:,-1].values
```

```python
#training a linear model
from sklearn.linear_model import LinearRegression
lin_reg = LinearRegression()
lin_reg.fit(X,y)
```

```python
#training poly model

from sklearn.preprocessing import PolynomialFeatures
poly_reg = PolynomialFeatures(degree=4)
x_poly=poly_reg.fit_transform(X)
lin_regr2= LinearRegression()
lin_regr2.fit(x_poly,y)
# print(x_poly)
```

```python
#visualising linear model

plt.scatter(X,y,color='red')
plt.plot(X, lin_reg.predict(X),color='blue')
plt.title('T or F(LR)')
plt.xlabel('Position Level')
plt.ylabel('Salary Level')
plt.show()
```

```python
#visualising poly 

plt.scatter(X,y,color='red')
plt.plot(X,lin_regr2.predict(poly_reg.fit_transform(X)))
plt.title('poly regr')
plt.xlabel('Positon salary')
plt.ylabel('Salary Level')
plt.show()
```

```python
#predicting new result with linear regr

lin_reg.predict([[6.5]])
```

```python
#predicting new result with poly regr

lin_regr2.predict(poly_reg.fit_transform([[6.5]]))
```
